```python
import os
import sys
import matplotlib.pyplot as plt
import csv


from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
```

```python
image_path  = "../../data/VOCdevkit/VOC2007/JPEGImages/000012.jpg"
positive_path = "../../data/finetune_car/train\Annotations/000012_1.csv"
negative_path = "../../data/finetune_car/train\Annotations/000012_0.csv"

with open(positive_path, 'r') as f:
    positive_list = csv.reader(f)
    positive_list = [p for p in positive_list]
    
with open(negative_path, 'r') as f:
    negative_list = csv.reader(f)
    negative_list = [n for n in negative_list]
    
    
print('positive_list', len(positive_list), positive_list[0][0])
print('negative_list', len(negative_list), negative_list[0])
```

```python

image = Image.open(image_path).convert('RGB')
image_positive = image.copy()
image_negative = image.copy()
draw = ImageDraw.Draw(image_positive)

image_num = 5
positive_box = positive_list[image_num][0].split(' ')
positive_point1 = (int(positive_box[0]), int(positive_box[2]))
positive_point2 = (int(positive_box[1]), int(positive_box[3]))

# negative_box = negative_list[0]

print('point1', positive_point1)
print('point2', positive_point2)

# Box를 그릴 때, 왼쪽 상단 점과, 오른쪽 하단 점의 좌표를 입력으로 주면 됩니다.
draw.rectangle((positive_point2, positive_point1),
                outline = "red",
                width = 2,
                )

plt.figure(figsize=(10,30))
# plt.show(image_positive)
plt.imshow(image_positive)

```
